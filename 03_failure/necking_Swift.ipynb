## Diffuse necking: Swift's model

### Uniaxial tension

Swift (1952) formulated a criterion to predict diffuse necking in metal sheets.
A simplified formulation of this model under uniaxial tension conditions states that necking is initiated
when the force reaches its maximum.

$$
dF_1 = d(\sigma_1 A_1) = 0
\quad \rightarrow \quad
\frac{d\sigma_1}{\sigma_1} = -\frac{dA_1}{A_1} = \frac{dl_1}{l_1} = d\varepsilon_1
\quad \rightarrow \quad
\boxed{ \frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} = 1 }
$$

If we further use a power law to deﬁne the stress-strain relationship in direction 1, the left-hand expression of the above equation becomes:
$$
\sigma_1 = K \varepsilon_1^n
\quad \rightarrow \quad
\frac{d\sigma_1}{d\varepsilon_1} = K\,n\,\varepsilon_1^{n-1} \quad \rightarrow \quad
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} = \frac{n}{\varepsilon_1}
$$

Then we have a strain-based forming limit criterion:
$$
\frac{n}{\varepsilon_1} = 1  \quad \rightarrow \quad
\boxed{ \varepsilon_1^* = n }
$$


### Generalized model

In its general formulation, the Swift criterion states that the condition for forming a diﬀuse neck is satisﬁed when the load reaches a maximum along both principal directions ($dF_1 = 0$ and $dF_2 = 0$), leading to the following simultaneous constraints:

$$
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} = 1
\qquad , \qquad
\frac{1}{\sigma_2} \frac{d\sigma_2}{d\varepsilon_2} = 1
$$

Substituting into these equations the flow rule in the form:
$$
d\varepsilon_i = d\overline\varepsilon \frac{\partial\overline\sigma}{\partial\sigma_i}
$$
we obtain:
$$
d\sigma_1 = \sigma_1 d\varepsilon_1 = \frac{\partial\overline\sigma}{\partial\sigma_1} \sigma_1 d\overline\varepsilon
\quad , \quad
d\sigma_2 = \sigma_2 d\varepsilon_2 = \frac{\partial\overline\sigma}{\partial\sigma_2} \sigma_2 d\overline\varepsilon
$$

For a yield function that increases isotropically $\overline\sigma(\sigma_1, \sigma_2) = \sigma_y(\overline\varepsilon)$, the rate of change of the yield function can be defined in two ways:
$$
d\overline\sigma = \frac{\partial\overline\sigma}{\partial\sigma_1} d\sigma_1 + \frac{\partial\overline\sigma}{\partial\sigma_2} d\sigma_2 = 
\frac{\partial\sigma_y}{\partial\overline\varepsilon} d\overline\varepsilon
$$

Substituting into this equation the above expression of $d\sigma_i$:
$$
\frac{d\overline\sigma}{d\overline\varepsilon } = 
\left( \frac{\partial\overline\sigma}{\partial\sigma_1} \right)^2 \sigma_1 + \left( \frac{\partial\overline\sigma}{\partial\sigma_2} \right)^2 \sigma_2 
$$

Using a Von Mises yield function under plane stress conditions:
$$
\overline\sigma = \sqrt{\sigma_1^2+\sigma_2^2-\sigma_1 \sigma_2}
$$
$$
\frac{\partial\overline\sigma}{\sigma_1} = \frac{2\sigma_1-\sigma_2}{2\sqrt{\sigma_1^2+\sigma_2^2-\sigma_1 \sigma_2}}
\quad , \quad
\frac{\partial\overline\sigma}{\sigma_2} = \frac{2\sigma_2-\sigma_1}{2\sqrt{\sigma_1^2+\sigma_2^2-\sigma_1 \sigma_2}}
$$
and substituting into the above equation:
$$
\frac{d\overline\sigma}{d\overline\varepsilon } =  
\frac{(2\sigma_1-\sigma_2)^2 \sigma_1 + (2\sigma_2-\sigma_1)^2 \sigma_2}{4(\sigma_1^2+\sigma_2^2-\sigma_1 \sigma_2)}
$$
$$
\boxed{ \frac{1}{\overline\sigma}\frac{d\overline\sigma}{d\overline\varepsilon } =  \frac{(2-\alpha)^2 + (2\alpha-1)^2 \alpha}{4(1 - \alpha + \alpha^2)^{3/2}} }
$$
where $\alpha = \sigma_2/\sigma_1$. Note that in unixial tension ($\alpha = 0$), the last expression reduces to:
$$
\frac{1}{\overline\sigma}\frac{d\overline\sigma}{d\overline\varepsilon } = \frac{1}{\sigma_1}\frac{d\sigma_1}{d\varepsilon_1 } = 1
$$

If we further use a power law to deﬁne the stress-strain relationship:
$$
\sigma_y = K \overline\varepsilon^n
\quad \rightarrow \quad
\frac{d\sigma_y}{d\overline\varepsilon} = K\,n\,\overline\varepsilon^{n-1} \quad \rightarrow \quad
\frac{1}{\sigma_y} \frac{d\sigma_y}{d\overline\varepsilon} = \frac{n}{\overline\varepsilon}
$$

Then we have a strain-based forming limit criterion:
$$
\frac{1}{\overline\sigma}\frac{d\overline\sigma}{d\overline\varepsilon } = 
\frac{1}{\sigma_y} \frac{d\sigma_y}{d\overline\varepsilon}
\quad \rightarrow \quad
\overline\varepsilon = n 
\frac{4(1 - \alpha + \alpha^2)^{3/2}}{(2-\alpha)^2 + (2\alpha-1)^2 \alpha}
$$

$$
\boxed{ \varepsilon_1^* = n \frac{\sqrt{3}}{2\sqrt{1+\beta+\beta^2}}
\frac{4(1 - \alpha + \alpha^2)^{3/2}}{(2-\alpha)^2 + (2\alpha-1)^2 \alpha} }
\quad , \quad
\boxed{ \varepsilon_2^* = \beta \, \varepsilon_1 }
$$
where $\alpha$ and $\beta$ are related from the flow rule:
$$
\alpha = \frac{\sigma_2}{\sigma_1} = \frac{2\beta+1}{2+\beta} \quad , \quad
\beta = \frac{d\varepsilon_2}{d\varepsilon_1} = \frac{2\alpha-1}{2-\alpha}
$$
