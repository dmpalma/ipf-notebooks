## Localized necking: Hill's model

Hill (1952) formulated a criterion to predict localized necking in metal sheets.
A simplified formulation of this model states that the localization is initiated
when the tension (force per unit of width) reaches its maximum.

$$
dT_1 = d(\sigma_1 t) = 0 \quad \rightarrow \quad
\frac{d\sigma_1}{\sigma_1} = -\frac{dt}{t} = -d\varepsilon_3 =(1+\beta)d\varepsilon_1
$$

$$
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} = 1 + \beta
$$

From the flow rule:
$$
\alpha = \frac{\sigma_2}{\sigma_1} = \frac{2\beta+1}{2+\beta} \quad , \quad
\beta = \frac{d\varepsilon_2}{d\varepsilon_1} = \frac{2\alpha-1}{2-\alpha}
$$

$$
\varphi = \frac{\overline\sigma}{\sigma_1} = \sqrt{1-\alpha+\alpha^2} \quad , \quad 
\rho = \frac{d\overline\varepsilon}{d\varepsilon_1} = \frac{2}{\sqrt{3}}\sqrt{1+\beta+\beta^2}
$$

### Hollomon's law

Assuming a Hollomon's strain-hardening law:
$$
\overline\sigma = K \overline\varepsilon^n \quad \rightarrow \quad
\sigma_1 = K' \varepsilon_1^n \quad , \quad
K' = \frac{K \rho^n}{\varphi}
$$

$$
\frac{d\sigma_1}{d\varepsilon_1} = K' n \varepsilon_1^{n-1} \quad \rightarrow \quad
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} = \frac{n}{\varepsilon_1}
$$

Therefore:
$$
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} = \frac{n}{\varepsilon_1} = 1+\beta  \quad \rightarrow \quad
\varepsilon_1^* = \frac{n}{1+\beta}
$$

### Swift's law

In general, metal sheets have undergone a prior deformation in the rolling process. Using the Swift's constitutive law:

$$
\overline\sigma = K (\varepsilon_0 + \overline\varepsilon)^n \quad\rightarrow\quad
\sigma_1 = K' (\varepsilon_0 + \rho \, \varepsilon_1)^n \quad , \quad
K' = \frac{K}{\varphi}
\quad\rightarrow\quad
$$
$$
\quad\rightarrow\quad
\frac{d\sigma_1}{d\varepsilon_1} = K' \, n (\varepsilon_0 + \rho \, \varepsilon_1)^{n-1} \, \rho \quad\rightarrow\quad
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} 
= \frac{K' \, n (\varepsilon_0 + \rho \, \varepsilon_1)^{n-1} \, \rho}{K' (\varepsilon_0 + \rho \, \varepsilon_1)^n} 
= \frac{n \, \rho}{\varepsilon_0 + \rho \, \varepsilon_1}
$$

Therefore:
$$
\frac{1}{\sigma_1} \frac{d\sigma_1}{d\varepsilon_1} =
\frac{n \, \rho}{\varepsilon_0 + \rho \, \varepsilon_1} = 1+\beta \quad\rightarrow\quad
\varepsilon_1^* = \frac{n}{1+\beta} - \frac{\varepsilon_0}{\rho}
$$
where $\rho$ is a function of $\beta$. Substituting:
$$
\varepsilon_1^* = \frac{n}{1+\beta} - \varepsilon_0 \frac{\sqrt{3}}{2}\frac{1}{\sqrt{1+\beta+\beta^2}}
$$

Notice that setting $\varepsilon_0 = 0$ in this expression, it reduces to the previous one obtained for the Hollomon's power law.
